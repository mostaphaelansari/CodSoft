```python
genre_matrix = pd.DataFrame(0, columns=genres, index=df.index)
```

```python
genre_keywords = {
    "Action": ["action", "thrilling", "adventure"],
    "Adventure": ["adventure", "exciting"],
    "Drama": ["drama", "emotional"],
    "Romance": ["romantic", "love"],
    "Comedy": ["comedy", "funny"],
    "Science Fiction": ["sci-fi", "future"]
}
```

```python
for genre, keywords in genre_keywords.items():
    for keyword in keywords:
        genre_matrix[genre] |= string.str.contains(keyword, case=False)
```
